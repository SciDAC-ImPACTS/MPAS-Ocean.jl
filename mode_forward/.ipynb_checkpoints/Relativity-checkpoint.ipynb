```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
```

```python
X0 = np.linspace(0, 100, 50)
T0 = np.linspace(0, 100, 50)
XT0, TX0 = np.meshgrid(X0, T0)
```

```python
s = np.linspace(-0.0j, 100.0j, 50)
```

```python
s
```

```python
p0x = 50*np.ones(s.shape)
p0t = np.sqrt( p0x**2 - s**2 )
```

```python
def spaceTimeGraph(ax, XT, TX, px, pt, L=lambda x,t : (x,t)):
    ax.scatter(*L(XT, TX), s=2)
    ax.plot(*L(px, pt), marker='s', color='orange')
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p0x, p0t)
```

$ s^2 = x^2 - t^2 $

$ t = \sqrt{x^2 - s^2} $

```python
p1x = np.linspace(20, 80, X0.shape[0])
p1t = np.sqrt( p1x**2 - s**2 + 0.0j )
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p1x, p1t)
```

```python
def makeLorentz(px, pt):
    v = np.mean( np.gradient(px) / np.gradient(pt) )
    lf = np.sqrt(1 - v**2)
    return lambda x, t : ( lf*(x-v*t), lf*(t-v*x) )
```

```python
L1 = makeLorentz(p1x, p1t)
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p1x, p1t, L=L1)
```

```python
p2x = np.cos(X0/10)*4 + X0*0.4 + 10
p2x = p2x/np.mean(np.gradient(p2x)) * 0.9 # normalize to sub-(0.9)light speed, by making each interval step less than 1 distance
p2t = np.sqrt( p2x**2 - s**2 + 0.0j )
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p2x, p2t)
```

```python
L2linear = makeLorentz(p2x, p2t)
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p2x, p2t, L=L2linear)
```

```python
def makeLorentzNL(px, pt):
    v = np.gradient(px) / np.gradient(pt)
    lf = np.sqrt(1 - v**2)
    return lambda x, t : ( lf*(x-v*t), lf*(t-v*x) )
```

```python
L2nl = makeLorentzNL(p2x, p2t)
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p2x, p2t, L=L2nl)
```

```python
p3x = (T0)**2 * 0.005 + 20
p3x = p3x/np.mean(np.gradient(p3x)) * 0.9 # normalize to sub-(0.9)light speed, by making each interval step less than 1 distance
p3t = np.sqrt( p3x**2 - s**2 + 0.0j )
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p3x, p3t)
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p3x, p3t, L=makeLorentz(p3x, p3t))
```

```python
fig, ax = plt.subplots(1)
spaceTimeGraph(ax, XT0, TX0, p3x, p3t, L=makeLorentzNL(p3x, p3t))
```

```python
fig, ax = plt.subplots(1)

def frame(i):
    px = (T0)**2 * 0.005 + i - 60
    px = px/np.mean(np.gradient(px)) * 0.9 # normalize to sub-(0.9)light speed, by making each interval step less than 1 distance
    pt = np.sqrt( px**2 - s**2 + 0.0j )
    spaceTimeGraph(ax, XT0, TX0, px, pt, L=makeLorentzNL(px, pt))
    return [ax]

anim = animation.FuncAnimation(fig, frame, frames=np.arange(0, 200, 1), interval=20, blit=True, repeat=True)
```
