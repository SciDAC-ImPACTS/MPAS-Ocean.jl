```julia
myMPAS_O = MPAS_Ocean(false,
                CODE_ROOT * "MPAS_O_Shallow_Water/MPAS_O_Shallow_Water_Mesh_Generation/InertiaGravityWaveMesh",
                "base_mesh.nc", "mesh.nc", periodicity="Periodic")
ocn_init_routines_compute_max_level!(myMPAS_O)
```

```julia
meanCoriolisParameterf = sum(myMPAS_O.fCell) / myMPAS_O.nCells
```

```julia
meanFluidThicknessH = sum(myMPAS_O.bottomDepth)/length(myMPAS_O.bottomDepth)
```

```julia
rossbyRadiusR = sqrt(gravity*meanFluidThicknessH)/meanCoriolisParameterf
```

$ \pmatrix{\eta, u, v} = \pmatrix{A, U, V} cos(kxx, kyy, wt) $

$ kx U + kx V + w A = 0 $

```julia
kx = 4 * 2*pi / myMPAS_O.lX
U = 2.0e0

ky = 3  * 2*pi / myMPAS_O.lY
V = 4.5e0

k = sqrt(kx^2 + ky^2)

w = sqrt(meanCoriolisParameterf^2 + gravity*meanFluidThicknessH*k^2)
A = (kx*U + ky*V) / w

kx*U + ky*V + w*A
```

```julia
function inertiaGravityInit!(myMPAS_O, t=0; kx=kx, U=U, ky=ky, V=V, w=w, A=A)
    for iCell in 1:myMPAS_O.nCells
        x = myMPAS_O.xCell[iCell]
        y = myMPAS_O.yCell[iCell]
        
        myMPAS_O.sshCurrent[iCell] = A*cos(kx*x + ky*y - w*t)
    end
    
    for iEdge in 1:myMPAS_O.nEdges
        x = myMPAS_O.xEdge[iEdge]
        y = myMPAS_O.yEdge[iEdge]
        
        u = U*cos(kx*x + ky*y - w*t)
        
        v = V*cos(kx*x + ky*y - w*t)
        
        theta = myMPAS_O.angleEdge[iEdge]
        
        myMPAS_O.normalVelocityCurrent[iEdge] = u*cos(theta) + v*sin(theta)
    end
end
```

```julia
inertiaGravityInit!(myMPAS_O)
```

```julia
fig, ax, cbar, _ = heatMapMesh(myMPAS_O, myMPAS_O.sshCurrent)
vectorStreamPlotMesh(myMPAS_O, myMPAS_O.normalVelocityCurrent, fig=fig, ax=ax, cbar=cbar, xbins=17, ybins=13)
display(fig)
```

```julia
fig, ax, _, = vectorStreamPlotMesh(myMPAS_O, myMPAS_O.normalVelocityCurrent)
display(fig)
```

## define dt based on cfl condition

```julia
dt = 0.1 * minimum(myMPAS_O.dcEdge) / sqrt(gravity*meanFluidThicknessH)
```

# Simulate wave

```julia
2*pi/(w*dt) # number of timesteps per cycle of wave
```

```julia
nFrames = 30

sshOverTime = zeros(Float64, (nFrames, myMPAS_O.nCells))

for i in 1:nFrames
    forwardBackward!(myMPAS_O, dt, dt*6)
    sshOverTime[i,:] .= myMPAS_O.sshCurrent[:]
end
```

## animate simulation

```julia
cMax = maximum(abs.(sshOverTime))

fig, ax, _, col = heatMapMesh(myMPAS_O, sshOverTime[1,:], cMin=-cMax, cMax=cMax)

function nextFrame(i)
    col.set_array(sshOverTime[i+1,:])
end

anim = animation.FuncAnimation(fig, nextFrame, frames=nFrames)
```

```julia
ipydisplay.HTML(anim.to_html5_video())
```
