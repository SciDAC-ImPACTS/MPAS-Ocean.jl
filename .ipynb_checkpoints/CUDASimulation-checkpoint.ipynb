# Define Kernel Functions to Time-Progress Variables on GPU at individual Cells/Edges

```julia
function stepNormalVelocity!(nEdges, normalVelocityNew, normalVelocity, ssh, cellsOnEdge, nEdgesOnEdge, edgesOnEdge,
                weightsOnEdge, fEdge, dcEdge, gravity, dt)
    iEdge = (blockIdx().x - 1) * blockDim().x + threadIdx().x
    if iEdge <= nEdges

        normalVelocityNew[iEdge] = normalVelocity[iEdge]

        # take gradient of ssh across edge
        normalVelocityNew[iEdge] += dt * gravity * ( ssh[cellsOnEdge[1,iEdge]] - ssh[cellsOnEdge[2,iEdge]] ) / dcEdge[iEdge] # gravity term

        # coriolis term
        for i = 1:nEdgesOnEdge[iEdge]
            normalVelocityNew[iEdge] += dt * weightsOnEdge[i,iEdge] * normalVelocity[edgesOnEdge[i,iEdge]] * fEdge[edgesOnEdge[i,iEdge]]
        end

        normalVelocity[iEdge] = normalVelocityNew[iEdge]
        
    end
    return
end
```

```julia
function stepSSH!(nCells, sshNew, ssh, normalVelocity, bottomDepth, nEdgesOnCell, edgesOnCell, cellsOnCell,
                areaCell, edgeSignOnCell, dvEdge, dt)
    iCell = (blockIdx().x - 1) * blockDim().x + threadIdx().x
    if iCell <= nCells

        sshNew[iCell] = ssh[iCell]

        # sum flux through each edge of cell
        for i in 1:nEdgesOnCell[iCell]
            edgeID = edgesOnCell[i,iCell]
            neighborCellID = cellsOnCell[i,iCell]

            if neighborCellID !== 0
                depth = ( bottomDepth[neighborCellID] + bottomDepth[iCell] ) / 2.0
            else
                depth = bottomDepth[iCell]
            end
            flux = depth * normalVelocity[edgeID]
            sshNew[iCell] += dt * flux * edgeSignOnCell[iCell,i] * dvEdge[edgeID] / areaCell[iCell]
        end

        ssh[iCell] = sshNew[iCell]
        
    end
    return
end
```

```julia
function forwardBackwardStepGPU!(myMPAS_O, dt)
    synchronize()
    blockSizeNormVel = Int(round(MAX_THREADS_PER_BLOCK*2/3))
    blockSizeSSH     = MAX_THREADS_PER_BLOCK - blockSizeNormVel
    @cuda blocks=cld(myMPAS_O.nEdges, blockSizeNormVel) threads=blockSizeNormVel stepNormalVelocity!(
                                                        myMPAS_O.nEdges,
                                                        myMPAS_O.normalVelocityNew,
                                                        myMPAS_O.normalVelocityCurrent,
                                                        myMPAS_O.sshCurrent,
                                                        myMPAS_O.cellsOnEdge,
                                                        myMPAS_O.nEdgesOnEdge,
                                                        myMPAS_O.edgesOnEdge,
                                                        myMPAS_O.weightsOnEdge,
                                                        myMPAS_O.fEdge,
                                                        myMPAS_O.dcEdge,
                                                        gravity, dt)
    synchronize()
    @cuda blocks=cld(myMPAS_O.nCells, blockSizeSSH) threads=blockSizeSSH stepSSH!(
                                                        myMPAS_O.nCells,
                                                        myMPAS_O.sshNew,
                                                        myMPAS_O.sshCurrent,
                                                        myMPAS_O.normalVelocityNew,
                                                        myMPAS_O.bottomDepth,
                                                        myMPAS_O.nEdgesOnCell,
                                                        myMPAS_O.edgesOnCell,
                                                        myMPAS_O.cellsOnCell,
                                                        myMPAS_O.areaCell,
                                                        myMPAS_O.edgeSignOnCell,
                                                        myMPAS_O.dvEdge,
                                                        dt)
    synchronize()
end
```

# Define Initial Condition

```julia
function gaussianInit!(myMPAS_O; sx=1/3, sy=1/3, mx=1/2, my=1/2)
    sigmax = myMPAS_O.lX*sx
    sigmay = myMPAS_O.lY*sy
    
    mux = myMPAS_O.lX*mx
    muy = myMPAS_O.lY*my
    
    for iCell in 1:myMPAS_O.nCells
        myMPAS_O.sshCurrent[iCell] = exp( - (myMPAS_O.xCell[iCell] - mux)^2 / sigmax^2 - (myMPAS_O.yCell[iCell] - muy)^2 / sigmay^2 )
    end
    
    myMPAS_O.normalVelocityCurrent .= 0
    
    return nothing
end
```

```julia
function planeWaveInit!(myMPAS_O; rkx = 1, rky = 1)
    kx = rkx * 2 * pi / myMPAS_O.lX
    ky = rky * 2 * pi / myMPAS_O.lY
    
    for iCell in 1:myMPAS_O.nCells
        myMPAS_O.sshCurrent[iCell] = cos( kx * myMPAS_O.xCell[iCell] + ky * myMPAS_O.yCell[iCell] )
    end
    
    myMPAS_O.normalVelocityCurrent .= 0
    
    myMPAS_O.sshNew[:] .= myMPAS_O.sshCurrent[:]
    myMPAS_O.normalVelocityNew[:] .= myMPAS_O.normalVelocityCurrent[:]
    
    return nothing
end
```

```julia
moveToDevice!(myMPAS_O, CPU())

planeWaveInit!(myMPAS_O) # could also do gaussianInit!
```

```julia
fig, ax, cbar, col = heatMapMesh(myMPAS_O, myMPAS_O.sshCurrent)
display(fig)
```

# Simulate Wave on GPU

```julia
# simulate and record nFrames snapshots of the SSH
nFrames = 30

sshOverTime = zeros(nFrames, myMPAS_O.nCells)

moveToDevice!(myMPAS_O, GPU())
for i in 1:nFrames
    for j in 1:100
        forwardBackwardStepGPU!(myMPAS_O, 0.1)
    end
    sshOverTime[i,:] = myMPAS_O.sshCurrent[:]
end
```

# Animate Simulation

```julia
# move it back to CPU for animation
moveToDevice!(myMPAS_O, CPU())

cMax = maximum(abs.(sshOverTime))

fig, ax, _, col = heatMapMesh(myMPAS_O, sshOverTime[1,:], cMin=-cMax, cMax=cMax)

function nextFrame(i)
    col.set_array(sshOverTime[i+1,:])
end

anim = animation.FuncAnimation(fig, nextFrame, frames=nFrames)
```

```julia
ipydisplay.HTML(anim.to_html5_video())
```

# Compare run time of GPU vs CPU versions

```julia
using BenchmarkTools
```

```julia
moveToDevice!(myMPAS_O, CPU())
gaussianInit!(myMPAS_O)
# save gaussian initial condition so it doesn't have to be recalculated every time
initialSSH = myMPAS_O.sshCurrent[:]
initialNormalVelocity = myMPAS_O.normalVelocityCurrent[:]
```

```julia
# benchmark time it takes to do a timestep on the GPU
myMPAS_O.sshCurrent[:] = initialSSH[:]
myMPAS_O.normalVelocityCurrent[:] = initialNormalVelocity[:]
moveToDevice!(myMPAS_O, GPU())
gpubench = @benchmark forwardBackwardStepGPU!(myMPAS_O, 0.1)
```

```julia
# benchmark time it takes to do a timestep on the CPU
myMPAS_O.sshCurrent[:] = initialSSH[:]
myMPAS_O.normalVelocityCurrent[:] = initialNormalVelocity[:]
moveToDevice!(myMPAS_O, CPU())
cpubench = @benchmark forwardBackwardSingleStep!(myMPAS_O, 0.1)
```

# Vectorized versions (broadcast syntax rather than for loops) of tendencies

```julia
function stepNormalVelocityVectorized!(normalVelocityNew, normalVelocity, ssh, cellsOnEdge, nEdgesOnEdge, edgesOnEdge, weightsOnEdge, fEdge, dcEdge, dt)
    normalVelocityNew[:] .= normalVelocity[:]
    
    # take gradient of sshCurrent across edge
    normalVelocityNew[:] .+= gravity .* ( ssh[cellsOnEdge[1,:]] .- ssh[cellsOnEdge[2,:]] ) ./ dcEdge
    # TODO boundary condition, will error when cell1Index == 0 or cell2Index == 0
    
    normalVelocity2 = map(iEdge -> iEdge != 0 ? normalVelocity[iEdge] : 0, edgesOnEdge)
    fEdge2 = map(iEdge -> iEdge != 0 ? fEdge[iEdge] : 0, edgesOnEdge)
    
    # coriolis term

    # edgesOnEdge is a 2x2 matrix, each value is a valid edge index
    # so any edge-shaped thing indexed with it should turn to its shape?
    normalVelocityNew[:] .+= dropdims(reduce(+, weightsOnEdge[:,:] .* normalVelocity2 .* fEdge2, dims=1), dims=1) # i, iEdge => iEdge
    # TODO boundary condition, will error when any(edgesOnEdge == 0)
    
    normalVelocity[:] = normalVelocityNew[:]
    
    return nothing
end
```

```julia
function stepSSHVectorized!(sshNew, ssh, normalVelocity, bottomDepth, nEdgesOnCell, edgesOnCell, cellsOnCell, areaCell, edgeSignOnCell, dvEdge, dt)
    
    sshNew[:] .= ssh[:]
    
    # sum flux through each edge of cell
    
    bottomDepthOnCellNeighbors = map(iCell -> iCell != 0 ? bottomDepth[iCell] : 0, cellsOnCell)
    normalVelocityOnCellEdges = map(iEdge -> iEdge != 0 ? normalVelocity[iEdge] : 0, edgesOnCell)
    dvEdgeOnCellEdges = map(iEdge -> iEdge != 0 ? dvEdge[iEdge] : 0, edgesOnCell)
    
    mean_depth = ( bottomDepthOnCellNeighbors .+ reshape(bottomDepth, (1, size(bottomDepth)...) ) ) ./ 2.0
    flux = mean_depth .* normalVelocityOnCellEdges
    sshNew[:] .+= dropdims(reduce(+, flux .* edgeSignOnCell .* dvEdgeOnCellEdges ./ reshape(areaCell, (1, size(areaCell)...)), dims=1), dims=1) # i, iCell => iCell
    # TODO: boundary conditions, will error currently when edgeID == 0
    
    ssh[:] = sshNew[:]
    
    return nothing
end
```

```julia
function forwardBackwardStepVectorized!(myMPAS_O, dt)
    
    stepNormalVelocityVectorized!(myMPAS_O.normalVelocityNew,
                                                        myMPAS_O.normalVelocityCurrent,
                                                        myMPAS_O.sshCurrent,
                                                        myMPAS_O.cellsOnEdge,
                                                        myMPAS_O.nEdgesOnEdge,
                                                        myMPAS_O.edgesOnEdge,
                                                        myMPAS_O.weightsOnEdge,
                                                        myMPAS_O.fEdge,
                                                        myMPAS_O.dcEdge,
                                                        dt)

    stepSSHVectorized!(myMPAS_O.sshNew,
                                             myMPAS_O.sshCurrent,
                                             myMPAS_O.normalVelocityNew,
                                             myMPAS_O.bottomDepth,
                                             myMPAS_O.nEdgesOnCell,
                                             myMPAS_O.edgesOnCell,
                                             myMPAS_O.cellsOnCell,
                                             myMPAS_O.areaCell,
                                             myMPAS_O.edgeSignOnCell,
                                             myMPAS_O.dvEdge,
                                             dt)
end
```

# Let CUDA automatically automatically find a way to run it on the GPU
## by putting CuArrays into the vectorized operators
The way CUDA does this may not be optimally effecient. Likely this will be much slower than writing custom kernels.

```julia

nFrames = 30

sshOverTime = zeros(nFrames, myMPAS_O.nCells)

moveToDevice!(myMPAS_O, GPU())
for i in 1:nFrames
    for j in 1:10
        forwardBackwardStepVectorized!(myMPAS_O, 1.0)
    end
    sshOverTime[i,:] = myMPAS_O.sshCurrent[:]
end
```
